```python
import numpy as np 
import imageio
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import imutils
import tensorflow as tf
```

```python
im1 = imageio.imread('/home/iba/Music/wfh/AWS Projects/Indian Number Plates/licensed_car200.jpeg')
```

```python
plt.imshow(im1)
plt.show()
```

```python
img = cv2.resize(im1, (1000,1000) )
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert to grey scale
```

```python
plt.imshow(gray)
plt.show()
```

```python
gray = cv2.bilateralFilter(gray, 13, 15, 15)
```

```python
plt.imshow(gray)
plt.show()
```

```python
edged = cv2.Canny(gray, 400, 0) #Perform Edge detection
```

```python
plt.imshow(edged)
plt.show()
```

```python
contours=cv2.findContours(edged.copy(),cv2.RETR_TREE,
                                            cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)
contours = sorted(contours,key=cv2.contourArea, reverse = True)[:10]
screenCnt = None
```

```python
for c in contours:
    # approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.018 * peri, True)
    # if our approximated contour has four points, then
    # we can assume that we have found our screen
    if len(approx) == 4:
        screenCnt = approx
        break
```

```python
# Masking the part other than the number plate
mask = np.zeros(gray.shape,np.uint8)
new_image = cv2.drawContours(mask,[screenCnt],0,255,-1,)
new_image = cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.imshow(new_image)
plt.show()
```

```python
# Now crop
(x, y) = np.where(mask == 255)
(topx, topy) = (np.min(x), np.min(y))
(bottomx, bottomy) = (np.max(x), np.max(y))
Cropped = gray[topx:bottomx+1, topy:bottomy+1]
```

```python
plt.imshow(Cropped)
plt.show()
```

```python
from sklearn import svm
```

```python
from sklearn.svm import SVC
```

```python
#Read the number plate
text = svm.SVC(Cropped, kernel="Radial Basis Function")
print("Detected license plate Number is:",text)

img = cv2.resize(img,(500,300))
Cropped = cv2.resize(Cropped,(400,200))
cv2.imshow('car',img)
cv2.imshow('Cropped',Cropped)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
